```python
import pandas as pd

# Create sample dataset
data = {
    "study_hours": [2, 4, 6, 8, 3, 7, 5, 9, 4, 6],
    "sleep_hours": [8, 7, 5, 4, 6, 3, 5, 4, 7, 5],
    "screen_time": [2, 4, 6, 8, 3, 7, 5, 9, 4, 6],
    "burnout": [0, 0, 1, 1, 0, 1, 1, 1, 0, 1]
}

df = pd.DataFrame(data)

df
```

```python
from sklearn.linear_model import LogisticRegression

# Features (inputs)
X = df[["study_hours", "sleep_hours", "screen_time"]]

# Target (output)
y = df["burnout"]

# Create model
model = LogisticRegression()

# Train model
model.fit(X, y)

print("Model trained successfully!")
```

```python
# Fix warning by using DataFrame with column names

new_data = pd.DataFrame({
    "study_hours": [7],
    "sleep_hours": [4],
    "screen_time": [6]
})

prediction = model.predict(new_data)

if prediction[0] == 1:
    print("Burnout Risk: HIGH ⚠️")
else:
    print("Burnout Risk: LOW ✅")
```

```python
import matplotlib.pyplot as plt

# Plot study hours vs burnout
plt.scatter(df["study_hours"], df["burnout"])

plt.xlabel("Study Hours")
plt.ylabel("Burnout Risk")
plt.title("Study Hours vs Burnout")

plt.show()
```

```python
# Save dataset as CSV file

df.to_csv("burnout_dataset.csv", index=False)

print("Dataset saved successfully!")
```

```python
import joblib

# Save trained model
joblib.dump(model, "burnout_model.pkl")

print("Model saved successfully!")
```

```python
# Load saved model

loaded_model = joblib.load("burnout_model.pkl")

# Test prediction again

test_data = pd.DataFrame({
    "study_hours": [6],
    "sleep_hours": [5],
    "screen_time": [7]
})

prediction = loaded_model.predict(test_data)

if prediction[0] == 1:
    print("Burnout Risk: HIGH ⚠️")
else:
    print("Burnout Risk: LOW ✅")
```
